```python
def grabImage(sid=1, imres=8):
# I will use the file name dbGrab.... for the image grabbed from the camera. First step is to get rid of previously grabbed images
    subprocess.run('rm dbGrab*', \
                   stdout=subprocess.PIPE, shell=True)
    subprocess.run('ssh -i ~/.ssh/id_rsa_cam.pub nvidia@10.236.176.38 "rm dbGrab*"', \
                   stdout=subprocess.PIPE, shell=True)

    # ok, now let's grab an image on the tegra
    subprocess.run('ssh -i ~/.ssh/id_rsa_cam.pub nvidia@10.236.176.38 "nvgstcapture-1.0 -m 1 --sensor-id '+str(sid)+' -A --capture-auto 1 \
                    --file-name dbGrab --image-res '+str(imres)+'"', stdout=subprocess.PIPE, shell=True)
    subprocess.run('scp -i ~/.ssh/id_rsa_cam.pub nvidia@10.236.176.38:dbGrab* .', \
                   stdout=subprocess.PIPE, shell=True)
    #ok, now let's find the file name of the image we just grabbed and transfered 
    pop=subprocess.run('ls',stdout=subprocess.PIPE, shell=True)
    rePop=pop.stdout
    rePop=rePop.decode("utf-8")
    fileName = re.search(r'dbGrab(.*).jpg', rePop)
    fileName=fileName.group()
    pop=imio.imread(fileName)
    pop=skimage.transform.rotate(pop,180)
#     fig=imio.imshow(pop)
#     fig.axes.get_xaxis().set_visible(False)
#     fig.axes.get_yaxis().set_visible(False)
#     strftime("%a, %d %b %Y %H:%M:%S", gmtime())
    return pop
```

```python
def focusMetric(ss=0):
    pip=grabImage(sid=ss, imres=8)
    pop=np.sum(np.abs(skimage.filters.laplace(pip[:,:,1])))
    return pop
```

```python
def grabImageExpo(sid=0, imres=8, expoTime=0.500):
# I will use the file name dbGrab.... for the image grabbed from the camera. First step is to get rid of previously grabbed images
    subprocess.run('rm dbGrab*', \
                   stdout=subprocess.PIPE, shell=True)
    subprocess.run('ssh -i ~/.ssh/id_rsa_cam.pub nvidia@10.236.176.38 "rm dbGrab*"', \
                   stdout=subprocess.PIPE, shell=True)

    # ok, now let's grab an image on the tegra
    subprocess.run('ssh -i ~/.ssh/id_rsa_cam.pub nvidia@10.236.176.38 "nvgstcapture-1.0 -m 1 --sensor-id '+str(sid)+' -A --capture-auto 1 \
                    --file-name dbGrab --exposure-time='+str(expoTime)+ '--image-res '+str(imres)+'"', stdout=subprocess.PIPE, shell=True)
    subprocess.run("scp -i ~/.ssh/id_rsa_cam.pub nvidia@10.236.176.38:dbGrab* .", \
                   stdout=subprocess.PIPE, shell=True)
    #ok, now let's find the file name of the image we just grabbed and transfered 
    pop=subprocess.run('ls',stdout=subprocess.PIPE, shell=True)
    rePop=pop.stdout
    rePop=rePop.decode("utf-8")
    fileName = re.search(r'dbGrab(.*).jpg', rePop)
    fileName=fileName.group()
    pop=imio.imread(fileName)
    pop=skimage.transform.rotate(pop,180)
#     fig=imio.imshow(pop)
#     fig.axes.get_xaxis().set_visible(False)
#     fig.axes.get_yaxis().set_visible(False)
#     strftime("%a, %d %b %Y %H:%M:%S", gmtime())
    return pop
```

```python
expo_dict = {}
for i in range(5):
    pop=grabImageExpo(expoTime = i*0.005+0.005)
    saturate_num = (pop > 0.9) + (pop < 0.1) # saturate: below 0.1 or beyond 0.9.
    saturate_num = saturate_num.astype(int)
    print(1-saturate_num.mean())
    expo_dict[str(i*0.005+0.005)] = 1-saturate_num.mean()
```

```python
print(expo_dict)
```

```python
bst_expo = float(max(expo_dict, key=expo_dict.get))
print("best expo time: {}".format(bst_expo))
pop=grabImageExpo(expoTime = bst_expo)
imio.imshow(pop)
imio.show()
```
