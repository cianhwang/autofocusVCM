```python
import numpy as np
import cv2
import random
import glob
import matplotlib.pyplot as plt
```

```python
image_size1 = 512
image_size2 = 512
Num_of_data = 10000
```

```python
Ground_truth_space = [500, 525, 550, 600, 650, 675, 725, 800, 875, 1000]
Data_space = list(range(1000, 450, -25))
Data_space = Data_space[::-1]
print("Data_space size:", len(Data_space))
```

```python
DATA = np.zeros([Num_of_data,image_size1,image_size2,1], dtype=np.uint8)
i = 0
Label = np.zeros([Num_of_data,2])
```

```python
def readImage(path):
    #print(path)
    fileName = glob.glob(path)[0]
    img = cv2.imread(fileName, 0)
    img = (img - img.min())/(img.max() - img.min())*255.0
    return img
```

```python
for i in range (Num_of_data//2):
    print(i)
    focus_position = random.choice(Ground_truth_space)
    focus_idx = Ground_truth_space.index(focus_position) + 4
    if focus_idx == 4:
        focus_idx = 3
    current_position = random.choice(Data_space)
    a = readImage('calibrate3/'+str(focus_idx)+'_'+str(focus_position)+'*.jpg')
    idx1= random.randint(-500,500) 
    idx2 =random.randint(-1000,1000) 
    DATA[2*i,:,:,0] = a[1080-256+idx1:1080+256+idx1,1920-256+idx2:1920+256+idx2].astype('uint8')
    Label[2*i,0] = focus_position
    Label[2*i,1] = focus_position
    
    a = readImage('calibrate3/'+str(focus_idx)+'_'+str(current_position)+'*.jpg')
    idx1= random.randint(-500,500) 
    idx2 =random.randint(-1000,1000) 
    DATA[2*i+1,:,:,0] = a[1080-256+idx1:1080+256+idx1,1920-256+idx2:1920+256+idx2].astype('uint8')
    Label[2*i+1,0] = focus_position
    Label[2*i+1,1] = current_position
    
```

```python
np.save('real_data_uint8_512_vcm_10000.npy',DATA)
np.save('real_label_uint8_512_vcm_10000.npy',Label)
```
